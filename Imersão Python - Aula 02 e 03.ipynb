```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/drive/MyDrive/Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel("/content/drive/MyDrive/Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/drive/MyDrive/Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/drive/MyDrive/Imersão Python - Tabela de ações.xlsx", sheet_name="Chatgpt")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_final', 'Var. Dia (%)':'Var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['Var_dia_pct'] / 100
df_principal['Valor_inicial'] = df_principal['Valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['Valor_final'] - df_principal['Valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}' .format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (em anos)':'Idade'}).copy()
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100 anos' if x > 100 else ('Menos de 50 anos' if x < 50 else 'Entre 50 e 100 anos'))
df_principal
```

```python
# Calculando o maior valor
maior = df_principal['Variacao_rs'].max()

# Calculando o menor valor
menor = df_principal['Variacao_rs'].min()

# Calculando a média
media = df_principal['Variacao_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
df_principal = df_principal.drop_duplicates()
df_principal
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
# Supondo que você tenha uma coluna 'Variacao_rs' em seu DataFrame df_analise_saldo
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação em Reais por Resultado')

# Formatação do texto para mostrar como moeda em reais (R$)
fig.update_traces(texttemplate='R$ %{text:.2f}', textposition='outside')

fig.show()
```

```python
# Construção do gráfico de pizza
fig = px.pie(df_analise_segmento, values='Variacao_rs', names='Segmento', title='Variação por Segmento')

# Converter os valores para reais
df_analise_segmento['Variacao_rs'] = df_analise_segmento['Variacao_rs'].map('R${:,.2f}'.format)

# Exibir o gráfico
fig.show()
```

```python
df_analise_cat_idade = df_principal.groupby('Cat_idade')['Variacao_rs'].sum().reset_index()
df_analise_cat_idade
```

```python
# Supondo que 'Cat_idade' seja a categoria que você está usando para agrupar
# 'Variacao_rs' seja a coluna de variação de receita
# 'Empresas' seja a coluna que representa o número de empresas

# Primeiro, vamos agrupar pelos valores de 'Cat_idade' e calcular a soma da 'Variacao_rs' e a contagem de 'Empresas'
df_analise_cat_idade = df_principal.groupby('Cat_idade').agg({'Variacao_rs': 'sum', 'Empresas': 'count'}).reset_index()

# Renomear as colunas para maior clareza
df_analise_cat_idade.columns = ['Cat_idade', 'Soma_Variacao_rs', 'Total_Empresas']

# Exibir o DataFrame resultante
print(df_analise_cat_idade)
```

```python
# Supondo que você tenha uma coluna 'Variacao_rs' em seu DataFrame df_analise_saldo
fig = px.bar(df_analise_cat_idade, x='Cat_idade', y='Variacao_rs', text='Variacao_rs', title='Variação em Reais por faixa etária')

# Formatação do texto para mostrar como moeda em reais (R$)
fig.update_traces(texttemplate='R$ %{text:.2f}', textposition='outside')

fig.show()
```
